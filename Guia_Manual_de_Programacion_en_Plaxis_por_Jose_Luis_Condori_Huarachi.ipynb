```python
#Importar Librerias
from plxscripting.easy import *
import numpy as np
import matplotlib.pyplot as plt
import time
import math
import csv
import pandas as pd
from datetime import datetime
from pathlib import Path
```

```python
# Conectar al servidor de Plaxis
s_i, g_i = new_server('localhost', 10000, password='tayusapa12A')
print("Conectado a PLAXIS")
```

```python
g_i.gotosoil()
# LISTA DE POLÍGONOS
# Cada polígono tiene 8 números: x1 y1 x2 y2 x3 y3 x4 y4
poligonos = [
    [0, 0, 100, 0, 100, 15, 0, 15],      # Polígono 1
    [0, 15, 100, 15, 100, 27, 0, 27],    # Polígono 2
    [0, 27, 100, 27, 100, 30, 0, 30]     # Polígono 3
]

# === CREACIÓN DE LOS POLÍGONOS ===
for i, coords in enumerate(poligonos, start=1):
    poly = g_i.polygon(*coords)[0]
    poly.Name = f"POLY_{i:02d}"
    print(f" Polígono {i} creado con coordenadas: {coords}")
```

```python
#Creacion de los Materiales
# MATERIAL 1 - Limo

Limo = g_i.soilmat()
g_i.setproperties(Limo,
    "Identification", "Limo",
    "SoilModel", "Hardening Soil",
    "DrainageType", "Drained",
    "gammaUnsat", 16.0,
    "gammaSat", 20.0,
    "E50Ref", 20000,
    "EOedRef",20000,
    "EURRef", 60000,
    "pRef", 100,
    "PowerM", 0.5,
    "cRef", 1,
    "phi", 30,
    "psi", 0,
    "nuUR", 0.2
)
print("Material Limo Creado.")

# MATERIAL 2 - Arena

Arena = g_i.soilmat()
g_i.setproperties(Arena,
    "Identification", "Arena",
    "SoilModel", "Hardening Soil",
    "DrainageType", "Drained",
    "gammaUnsat", 17.0,
    "gammaSat", 20.0,
    "E50Ref", 30000,
    "EOedRef", 30000,
    "EURRef", 90000,
    "pRef", 100,
    "PowerM", 0.5,
    "cRef", 0,
    "phi", 34,
    "psi", 4,
    "nuUR", 0.2
)
print("Material Arena Creado.")

# MATERIAL 3 - Franco

Franco = g_i.soilmat()
g_i.setproperties(Franco,
    "Identification", "Franco",
    "SoilModel", "Hardening Soil",
    "DrainageType", "Drained",
    "gammaUnsat", 17.0,
    "gammaSat", 19.0,
    "E50Ref", 12000,
    "EOedRef", 8000,
    "EURRef", 36000,
    "pRef", 100,
    "PowerM", 0.8,
    "cRef", 5,
    "phi", 29,
    "psi", 0,
    "nuUR", 0.2
)
print("Material Franco Creado.")
```

```python
#Asignacion de Materiales
g_i.Soils[0].Material = Franco
g_i.Soils[1].Material = Arena
g_i.Soils[2].Material = Limo
print("Materiales asignados a los poligonos del suelo")
```

```python
#Parametros del Muro
EA = 12000000 #Kn/m
EI = 120000 #Kn m2/m
w  = 8.3 # PesoKn/m/m
nu_val = 0.15 #Poisson

# Crear los Materiales
g_i.gotostructures()
pmat = g_i.platemat()
g_i.setproperties(pmat,
    "Identification", "PMat_Muro",
    "MaterialType",   "Elastic",
    "EA1",            EA,
    "EI",             EI,
    "w",              w,
    "StructNu",       nu_val
)
print("Material del Muro Creado")

# Dibujar los Muros
# Muro 1
x1a, y1a = 40, 30
x2a, y2a = 40, 14
plate1 = g_i.plate(x1a, y1a, x2a, y2a)[0]
plate1.Name = "Muro_1"

# Muro 2
x1b, y1b = 60, 30
x2b, y2b = 60, 14
plate2 = g_i.plate(x1b, y1b, x2b, y2b)[0]
plate2.Name = "Muro_2"

print("Muros Dibujados")

# Asignar Materiales
plates_target = list(g_i.Plates)[-2:]
for pl in plates_target:

        g_i.setmaterial((pl,), pmat)

print("Materiales Asignados")

```

```python
#Parametros
g_i.gotostructures()
E  = 210000000          #  (Kn/m2)
gamma  = 25            #  (kN/m³)
diam   = 0.105             #  (m)
spacing = 2.0             # (m)
#Crear Materiales
emat = g_i.embeddedbeammat()
g_i.setproperties(emat,
    "Identification", "EMat_Anclaje1",
    "MaterialType", "Elastic",
    "gamma", gamma,
    "Lspacing", spacing,
    "PredefinedCrossSectionType", "Solid circular beam",
    "Diameter", diam,
    "E", E ,
    "AxialSkinResistance", "Layer dependent"
)
print("Material Anclaje Creado.")

#Dibujar Materiales
coords = [
    (40, 27, 31, 21),
    (60, 27, 69, 21),
    (40, 23, 31, 17),
    (60, 23, 69, 17),
]

creados = []
for i, (x1, y1, x2, y2) in enumerate(coords, start=1):
    beam = g_i.embeddedbeam(x1, y1, x2, y2)[0]
    beam.Name = f"Anclaje_{i}"
    creados.append(beam)
print("Anclajes Dibujados.")

# Asignar Materiales
targets = list(g_i.EmbeddedBeams)[-4:]
for b in targets:
        g_i.setmaterial((b,), emat)

print("Materiales Asignados.")
```

```python
#Linea de Excavacion
g_i.gotostructures()
linea = g_i.line(40, 23, 60, 23)[0]
linea.Name = "Separador_1"
print("Linea Creada.")
```

```python
#Creacion de Carga 1
g_i.gotostructures()
#Coordenadas
x1, y1 = 28, 30
x2, y2 = 38, 30

ld=g_i.lineload(x1, y1, x2, y2)[0]
ld.Name = "Line_Load_1"

for carga in g_i.LineLoads:
    nombre = carga.Name.value

    if nombre == "LineLoad_1":
        carga.qy_start = -10
print("Carga Creada")
```

```python
#Creacion de Carga 2
g_i.gotostructures()
#Coordenadas
x1, y1 = 62.0, 30.0
x2, y2 = 72.0, 30.0

ld=g_i.lineload(x1, y1, x2, y2)[0]
ld.Name = "Line_Load_2"

for carga in g_i.LineLoads:
    nombre = carga.Name.value

    if nombre == "LineLoad_2":
        carga.qy_start = -10
print("Carga Creada")
```

```python
# Crear la Malla
g_i.gotomesh()
# 2. Establecer la calidad de la malla (Medium, Fine, Coarse.)
g_i.commands("meshcoarseness Medium")
g_i.mesh()
print("Malla Creada")
```

```python
#Apartado de Fases
g_i.gotostages()
```

```python
#Fase Inicial
fase_inicial = g_i.Phases[0]
for soil in g_i.Soils:
    g_i.activate(soil, g_i.InitialPhase)
print("Suelos Activados")
```

```python
#Fase de Muros y Cargas
fase_conf = g_i.phase(g_i.InitialPhase)
fase_conf.Identification = "Muro_Carga"

#Activar las Cargas
for carga in g_i.LineLoads:
    g_i.activate(carga, fase_conf)

# Activar los Plates
for pl in g_i.Plates:
    g_i.activate(pl, fase_conf)
print("Muros y Cargas Activados")
```

```python
#Fase de Excavacion
base = g_i.Phases[-1]
fase_exc = g_i.phase(base)
fase_exc.Identification = "Excavacion_1"

# Elementos a Desactivar
objetivos = {"Soil_3_2"}

# Desactivar Suelo
for s in g_i.Soils:
    nombre = str(s.Name.value)
    if nombre in objetivos:
        g_i.deactivate(s, fase_exc)

print("Excavacion 1 Desactivada")
```

```python
# Fase de Anclaje
base = g_i.Phases[-1]
fase_anc = g_i.phase(base)
fase_anc.Identification = "Anclajes_1"
# Elementos a Activar
nombres_emb    = {"EmbeddedBeam_1_1", "EmbeddedBeam_2_1"}
# Activar Anclajes
for eb in g_i.EmbeddedBeams:
    if str(eb.Name.value) in nombres_emb:
        g_i.activate(eb, fase_anc)

print("Anclaje 1 Activados ")
```

```python
#Fase de Excavacion
base = g_i.Phases[-1]
fase_exc = g_i.phase(base)
fase_exc.Identification = "Excavacion_2"

# Elementos a Desactivar
objetivos = {"Soil_2_1"}

# Desactivar Suelo
for s in g_i.Soils:
    nombre = str(s.Name.value)
    if nombre in objetivos:
        g_i.deactivate(s, fase_exc)

print("Excavacion 2 Desactivada")
```

```python
# Fase de Anclaje
base = g_i.Phases[-1]
fase_anc = g_i.phase(base)
fase_anc.Identification = "Anclajes_2"

# Elementos a Activar

nombres_emb    = {"EmbeddedBeam_3_1", "EmbeddedBeam_4_1"}

# Activar Anclajes
for eb in g_i.EmbeddedBeams:
    if str(eb.Name.value) in nombres_emb:
        g_i.activate(eb, fase_anc)

print("Anclaje 2 Activados ")
```

```python
#Fase de Excavacion
base = g_i.Phases[-1]
fase_exc = g_i.phase(base)
fase_exc.Identification = "Excavacion_3"

# Elementos a Desactivar
objetivos = { "Soil_2_2"}

# Desactivar Suelo
for s in g_i.Soils:
    nombre = str(s.Name.value)
    if nombre in objetivos:
        g_i.deactivate(s, fase_exc)

print("Excavacion 3 Desactivada")
```

```python
#Fase de Factor de Seguridad
base = g_i.Phases[-1]
fase_fs = g_i.phase(base)
fase_fs.Identification = "Factor de Segurida"

# Cambiar Tipo de Calculo
g_i.setproperties(fase_fs, "DeformCalcType", "Safety")

print("Factor de seguridad activado")
```

```python
#Ejecucion de Calculo
g_i.calculate()
print("Calculo Realizado ")
```
